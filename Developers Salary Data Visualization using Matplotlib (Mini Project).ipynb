```python
# VERY IMPORTANT COUNTER FUNCTION
# USE CASE IN NUMPY

from collections import Counter

with open(r'C:\Users\HELLO\Downloads\matp_data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_counter)
print('\n')
print(language_counter.most_common(10))
```

```python
language = []
popularity = []

for items in language_counter.most_common(10):
    language.append(items[0])
    popularity.append(items[1])
```

```python
# pyplot.bar(language, popularity)

pyplot.title('Most Popular Programming Language')
pyplot.ylabel('Language')
pyplot.xlabel('Total Users')

# It is not readable in vertical bar chart instead we should use HORIZONTAL BAR CHART

language.reverse()
popularity.reverse()    
# used it to make bar chart bottom down(optional)

pyplot.barh(language, popularity)
```

```python
# AREA FILLED LINE CHARTS

pyplot.style.use('fivethirtyeight')

pyplot.plot(dev_x,all_dev_y, label = 'All Types of Developers', linestyle = '--', color = 'black', linewidth = 1)
pyplot.plot(dev_x,py_dev_y, label = 'Python Developers', color = 'black', linewidth = 1)

pyplot.fill_between(dev_x,py_dev_y,all_dev_y,
                   where = (py_dev_y > all_dev_y),  alpha = 0.25, label = 'above average')

pyplot.fill_between(dev_x,py_dev_y, all_dev_y,
                   where = (py_dev_y <= all_dev_y),color = 'red' , alpha = 0.25, label = 'below average')



pyplot.title('Median Salary of Developers by Age')
pyplot.xlabel('Age')
pyplot.ylabel('Median Salary (USD)')
pyplot.tight_layout()

pyplot.legend()
pyplot.show()
```
